```python
# Connect to the ONTIME_REPORTING_01 CSV file
flight_data1 = "Resources/ONTIME_REPORTING_01_TC.csv"

# Create the airport_df DataFrame
flight_df = pd.read_csv(flight_data1)
```

```python
# Display the flight_df DataFrame
flight_df 
```

```python
# Keep the columns that are relevant to the machine learning model
columns = ["MONTH", "DAY_OF_MONTH", "DATE", "DAY_OF_WEEK","ORIGIN_AIRPORT_ID","DEST_AIRPORT_ID","DEST_CITY_NAME","DEP_TIME","DEP_DELAY_NEW","ARR_TIME","ARR_DELAY_NEW","DEP_DEL15","CANCELLED","DISTANCE","DISTANCE_GROUP"]
flight_df2 = flight_df[columns]
flight_df2
```

```python
# Determine the unique values of the "CANCELLED" flight column
flight_df2["CANCELLED"].unique().tolist()
```

```python
# Drop flights (rows) that are cancelled instead of dealyed
flight_df2 = flight_df2.drop(flight_df2[flight_df2["CANCELLED"] == 1.0].index)
flight_df2

#dataframe.drop(dataframe[<dataframe[column] == <value>>].index)
```

```python
# Check if there are null values in the flight_df2 DataFrame
flight_df2.isnull().sum()
```

```python
# Drop flights (rows) that contains null value(s) in the DataFrame
flight_df2 = flight_df2.dropna()

flight_df2
```

```python
# Confirm there are no null values in the DataFrame
flight_df2.isnull().sum()
```

```python
# Determine the data types of the flight_df2 DataFrame
flight_df2.dtypes
```

```python
# Convert the float64 data type to int64
flight_df_cleaned = flight_df2.astype({"DEP_TIME": 'int', "DEP_DELAY_NEW": 'int', "ARR_TIME": 'int', "ARR_DELAY_NEW": 'int', "DEP_DEL15": 'int', "CANCELLED": 'int', "DISTANCE": 'int'})
```

```python
# Confirm the data types of the flight_df_cleaned DataFrame
flight_df_cleaned.dtypes
```

```python
# Merge the 2 DataFrames (flight_df_cleaned and airport_df) 
merged_df = pd.merge(flight_df_cleaned, airport_df,on="ORIGIN_AIRPORT_ID",how='inner')
merged_df
```

```python
# Drop irrelevant columns
merged_df_cleaned = merged_df.drop(["DISPLAY_AIRPORT_NAME", "ORIGIN_CITY_NAME"], axis=1)
```

```python
# Dsiplay merged_df_cleaned DataFrame
merged_df_cleaned
```
