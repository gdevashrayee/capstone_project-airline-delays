```python
import pandas as pd

airport_df = pd.read_csv('airports_list.csv')
```

```python
airport_df.tail()
```

```python
ontime_01_df = pd.read_csv('ONTIME_REPORTING_01.csv')
ontime_01_df.head()
```

```python
ontime_02_df = pd.read_csv('ONTIME_REPORTING_04.csv')
print(ontime_02_df)
```

```python
ontime_03_df = pd.read_csv('ONTIME_REPORTING_07.csv')
print(ontime_03_df)
```

```python
ontime_04_df = pd.read_csv('ONTIME_REPORTING_10.csv')
print(ontime_04_df)
```

```python
airport_weather = pd.read_csv('Airport_Weather_2019.csv')
print(airport_weather)
```

```python
df_ontime_reporting=pd.concat([ontime_01_df,ontime_02_df,ontime_03_df,ontime_04_df],axis=0,join="outer")
df_ontime_reporting.tail()
```

```python
##df_ontime_reporting.to_csv('ontime_reporting.csv', index=False)
```

```python
columns = ["MONTH", "DAY_OF_MONTH","DAY_OF_WEEK","ORIGIN_AIRPORT_ID","DEST_AIRPORT_ID","DEST_CITY_NAME","DEP_TIME","DEP_DELAY_NEW","ARR_TIME","ARR_DELAY_NEW","DEP_DEL15","CANCELLED","DISTANCE","DISTANCE_GROUP"]
df_ontime_reporting_final=df_ontime_reporting[columns]
df_ontime_reporting_final.head()
```

```python
df_airport_reporting=pd.merge(airport_df,df_ontime_reporting_final,on='ORIGIN_AIRPORT_ID',how='outer')
df_airport_reporting.head()
```

```python
weather_df = pd.read_csv('Airport_weather_2019.csv')
weather_df.tail()
```

```python
df_airport_weather=pd.merge(airport_df,weather_df,on='NAME',how='inner')
df_airport_weather.tail()
```

```python
df_final=pd.merge(df_airport_reporting,df_airport_weather,on='NAME',how='inner')
df_final.tail()
```
